# Car Price Prediction

```python
#Import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
```

```python
#Input data
data = pd.read_csv('CarPrice_Assignment.csv')
```

```python
#Divide dependent and independent variable

predict = "price"
data = data[["symboling", "wheelbase", "carlength", "carwidth", "carheight", "curbweight", "enginesize", "boreratio", "stroke", "compressionratio", "horsepower", "peakrpm", "citympg", "highwaympg", "price"]]
x = np.array(data.drop([predict], 1))
y = np.array(data[predict])

print("Independent Dataset: \n", x)
print("\nDependent Dataset: \n", y)
```

```python
#Split the data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3,random_state=4)
```

```python
#Model selection and training
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
#Model evaluation
print("\nCoefficient of determination : \n",model.coef_)
print("\nMean squared error : \n",mean_squared_error(ytest,ypred))
print("\nR squared error : \n",r2_score(ytest,ypred))
```
